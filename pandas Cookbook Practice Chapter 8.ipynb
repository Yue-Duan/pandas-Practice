```python
import pandas as pd
```

##Unix timestamps

```python
popcon = pd.read_csv('popularity-contest.txt', sep=' ', )[:-1]
popcon.columns = ['atime', 'ctime', 'package-name', 'mru-program', 'tag']
popcon[:5]
```

Convert the numbers into integer, then time.  
unit could be in s, ms, ns

```python
popcon['atime'] = popcon['atime'].astype(int)
popcon['ctime'] = popcon['ctime'].astype(int)
popcon['atime'] = pd.to_datetime(popcon['atime'], unit='s')
popcon['ctime'] = pd.to_datetime(popcon['ctime'], unit='s')
```

```python
popcon['atime'].dtype
```

```python
popcon[:5]
```

##Find out packages that are not libraries

Get rid of rows with time=0. We can just use a string in this comparison, even though it's actually a timestamp on the inside.

```python
popcon = popcon[popcon['atime'] > '1970-01-01']
```

Use string operation to filter:

```python
nonlibraries = popcon[~popcon['package-name'].str.contains('lib')]
nonlibraries.sort('ctime', ascending=False)
```

This shows the packages that have benn installed latedly.
